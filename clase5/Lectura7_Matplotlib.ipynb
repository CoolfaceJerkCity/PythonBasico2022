```python
#Crea un programa que permita modificar las ejes cartesianos (axes-ejes)

from matplotlib.pylab import subplots
fig,ax = subplots()
x = np.linspace(-1,1,100)
y = x**3+2
ax.plot(x,y)
ax.set_xlabel('Tiempo')
ax.set_ylabel('Velocidad')
ax.set_title("Tiempo vs Velocidad")
ax.grid()
```

## Problema 3

```python
#En una figura con un solo subplot dibujar una traza senoidal y una cosenoidal
from matplotlib.pylab import subplots
fig,ax = subplots(figsize = (20,9))
x = np.linspace(-np.pi,np.pi,30)
y = np.sin(x)
y2 = np.cos(x)
ax.plot(x,y) #seno
ax.plot(x,y2,'or--') #Dotted... punteada
ax.set_xlabel('ángulo')
ax.set_ylabel('Sen/Cos')
ax.set_title("Seno y Coseno")
ax.legend(('Seno','Coseno'),loc = 'best') #best, upper right, lower left, right, center left, center right, lower center, upper center, center
ax.grid()
```

##Problema 4

```python
# Hacer 2 trazas, una del seno y otra del coseno de x, en una figura separadas. 
from matplotlib.pylab import subplots
fig,ax = subplots(2,1,figsize = (10,8))
x = np.linspace(-np.pi,np.pi,30)
y = np.sin(x)
y2 = np.cos(x)
ax[0].plot(x,y) #seno
ax[1].plot(x,y2,'or--') #Dotted... punteada
ax[0].set_xlabel('ángulo')
ax[0].set_ylabel('Sen/Cos')
ax[0].set_title("Seno y Coseno")
ax[0].grid()
```

## Problema 5

```python
#Cambiar el tipo de ejes para el sen de x
from matplotlib.pylab import subplots
fig,ax = subplots(figsize = (10,8))
x = np.linspace(-np.pi,np.pi,30)
y = np.sin(x)
ax.plot(x,y)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.grid()
plt.show()
```

```python

#Dibuja un parche de circunferencia
from matplotlib.patches import Circle
from matplotlib.pylab import subplots
fig2,axs2 = subplots()
axs2.add_patch(Circle((0,0),1,color = 'g',alpha = 0.9))
axs2.set_xlim(0,1.5)
axs2.set_ylim(-1,0)
axs2.grid(True)
```
